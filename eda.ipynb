분야는 그냥 dropna를 해야할듯.\
이건 가상의 값으로 대체하기가 힘들다.

```python
df = data.copy()
```

```python
df['기업가치'] = np.where(df['기업가치(백억원)'] == '6000이상', 7,
                np.where(df['기업가치(백억원)'] == '4500-6000', 6,
                np.where(df['기업가치(백억원)'] == '3500-4500', 5,
                np.where(df['기업가치(백억원)'] == '1500-2500', 3, 4))))
```

```python
df.isna().sum()
```

```python
df['기업가치'].value_counts()
```

기업가치만 다른 값으로 채워봤더니 데이터 불균형이 심해졌다.\
일단 dropna를 해보고 2000개 이하가 되면 다른 값으로 채워넣어야 할듯.

```python
df_cc = df.copy()
```

```python
df_c_dropna = df_cc.dropna(axis = 0)
```

```python
df_c_dropna
```

dropna를 했을 때 1800개 정도가 사라졌다.\
일단 이 데이터로 분석을 해봐야겠다.

```python
df_c_dropna['기업가치'].value_counts()
```
